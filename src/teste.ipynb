```python
# Selecionar apenas pessoas com toc e entre 30 e 60
df_toc = df[(df["Q11009"] == 1) & (df["C008"] >= 30)& (df["C008"] <= 60) ]
df_toc
```

```python

cols = ["Q00201","Q03001","Q060","Q06306","Q068",
        "Q074","Q079","Q088","Q092","Q11006","Q11604","Q120","Q128"]

df_bem = df[df[cols].eq(2).all(axis=1)]

df_bem = df_bem[(df_bem["C008"] >= 30) & (df_bem["C008"] <= 60)]

```

```python
df = pd.concat([df_toc, df_bem], ignore_index=True)
df
```
